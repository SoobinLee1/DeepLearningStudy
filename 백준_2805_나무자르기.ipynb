전형적인 이분탐색 알고리즘이다.

처음에 pivot을 while문 시작하면서 계산하게 했는데,이 것에 대한 반례로 1 2 2 나무 가 있을 경우를 들 수 있다.

image.png

```
N,M = list(map(int,input().split()))
trees = list(map(int,input().split()))
trees.sort()
left, right = 0,trees[-1]
pivot = 0
while left<=right:
  sum = 0
  for tree in trees:
    sum+=tree-pivot if tree-pivot>0 else 0
  if sum<M:
    right = pivot-1
    pivot = (left+right)//2
  elif sum>M:
    left = pivot+1
    pivot = (left+right)//2
  elif sum==M:
    break
print(pivot)
```
